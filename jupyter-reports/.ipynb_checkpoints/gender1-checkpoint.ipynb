```python
# Pandas
import pandas as pd
# Numpy
import numpy as np 
# Pyplot is a state-based interface to matplotlib. It provides a MATLAB-like way of plotting.
import matplotlib.pyplot as plt
# Display plots that are the output of running code cells  
%matplotlib inline
```

```python
# CSV filename
sheet_filename = "Data_Extract_From_Gender_Statistics/1746a01a-8d66-4c92-956d-a5097328a0d2_Data.csv"

# A comma-separated values (csv) file is returned 
#  as two-dimensional data structure with labeled axes.
#  df is a DataFrame
#
# sep: Delimiter to use
# index_col: Column(s) to use as the row labels of the DataFrame, 
#    either given as string name or column index. 
# encoding: Encoding to use for UTF when reading/writing (ex. ‘utf-8’).
# decimal: Character to recognize as decimal point (e.g. use ‘,’ for European data).
df = pd.read_csv(sheet_filename, sep=',', index_col=0, encoding='utf-8', decimal='.')
```

```python
# Return the first 3 rows (default of 5).
df.head(n=3)
```

```python
# Return the last 7 rows (default of 5).
df.tail(n=7)

# Observe that the last 5 rows we will cut in order to plot graphs.
```

```python
# This method prints information about a 
#  DataFrame including the index dtype and column dtypes, 
#  non-null values and memory usage.
df.info()
```

```python
# Generate descriptive statistics.
df.describe(include="all").T
```

```python
# Total of null values.

#  .isnull : Mask of bool values for each element in DataFrame that indicates 
#    whether an element is not an NA value.
#
#  .values : Return a Numpy representation of the DataFrame.
#
#  .sum : Return the sum of the array elements 
print(df.isnull().values.sum())
```

```python
# List of all columns in DataFrame.
lst_col = list(df.columns.values)

# Print the column list.
for i in range(len(lst_col)):
    print('Column {}: {}'.format(i, lst_col[i]))
```

```python
# Histogram for all countries in 2012

# Sequence of the command:
# .iloc[0:263] 
#     Remove last 5 rows.
# ['2012 [YR2012]']
#     Select column 2012 [YR2012]
# .apply(to_num)
#     Apply change to numeric data using the funciont to_num
# hist
#     Plot the data as histogram.


def to_num(v):
    try:
        return float(v)
    except:
        return float('nan')

plt_title='Women who believe a husband is justified in beating'\
    +' his wife when she refuses sex with him (%) in 2012 by countries'
df.iloc[0:263]['2012 [YR2012]'].apply(to_num).hist(bins=20)
plt.title(plt_title)
plt.xlabel('%')
plt.ylabel('Frequency by countries')
plt.xticks(np.arange(0, 100, step=5)) 
```

```python
# Histograms for all countries from 1990 to 2019

# Sequence of the command:
# .iloc[0:263] 
#     Remove last 5 rows.
# ['2012 [YR2012]']
#     Select column 2012 [YR2012]
# .apply(to_num)
#     Apply change to numeric data using the funciont to_num
# hist
#     Plot the data as histogram.


def to_num(v):
    try:
        return float(v)
    except:
        return float('nan')

count_figure = 1
# Get from 3 to end because the first 3 Columns are
#   Series Code, Country Name and Country Code
for year in list(df.columns.values)[3:]:
    plt.figure(count_figure)
    count_figure += 1
    
    plt_title='Women who believe a husband is justified in beating'\
        +' his wife when she refuses sex with him (%) in {} by countries'.format(year[:5])
    
    df.iloc[0:263][year].apply(to_num).hist(bins=20)
    plt.title(plt_title)
    plt.xlabel('%')
    plt.ylabel('Frequency by countries')
    plt.xticks(np.arange(0, 100, step=5)) 
```

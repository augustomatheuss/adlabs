## str_to_datetime(label)
    * Function used in **.groupby**, to transform the label removing the day.
## .iloc
    * Purely integer-location based indexing for selection by position.
## .to_datetime(str, format='%Y-%m-%d', errors='ignore')
    * Return datetime.
    * str: the string
    * format: datetime format
    * errors='ignore': invalid parsing will return the input
    
## .groupby
    * A groupby operation involves some combination of splitting the object, applying a function, and combining the results.

# Line plot of the mean of groups year-month X 'Close'. 

````python
def str_to_datetime(label):
    remove_day = str(pd.to_datetime(label, format='%Y-%m-%d', errors='ignore').year)\
    + '-' + '%02d' % pd.to_datetime(label, format='%Y-%m-%d', errors='ignore').month
    
    return remove_day


df.groupby(str_to_datetime)['Close'].mean().plot()
plt.title('Months X Mean of Close')
df.groupby(str_to_datetime).mean().iloc[0:]
````

## .iloc
    * Purely integer-location based indexing for selection by position.
    
## to_status(value)
    * Used in apply to transform values in 'HIGH', 'LOW' or 'EQUAL' based on the last Close value.
## hist
    * Plot the data as histogram.
    * figsize
        The size in inches of the figure to create. Uses the value in matplotlib.rcParams by default.

# Converting Close Column into HIGH, LOW or EQUAL related to the last Close value.

````python
last_close = df.iloc[0]['Close']
print('Last close value {}\n'.format(last_close))

def to_status(value):
    global last_close
    
    if value < last_close:
        return 'LOW'
    elif value > last_close:
        return 'HIGH'
    else:
        return 'EQUAL'
    
df['Close'].apply(to_status).hist(figsize=(3,5))
````

# Before merge, the dates must correspond.
    * The row labels in df_usd2eur must be converted to the same format in df
# Not all rows exist in both DataFrames
    * There are 464 rows in 'df' and 690 in 'df_usd2eur', because some days are missing in 'df' related to all days in the interval. This "missing" data will be filled with zeros after the merge.
    
## .index.values
    * ndarray with the indexes (row labels) of the DataFrame.
    
## 

````python
# Changes in rows_labels will change the indexes in DataFrame df_usd2eur.
rows_labels = df_usd2eur.index.values 

# Condition to not change twice the indexes
# The indexes of df_usd2eur are changed to the same format of the indexes in df DataFrame.
if len(rows_labels[0].split(" ")) == 2:
    for idx in range(rows_labels.size):
        # Example: 6/6/2018 23:58:00 -> 6/6/2018
        date_only = rows_labels[idx].split(" ")[0]
        # Example: 6/6/2018 -> ["6", "6", "2018"]
        date_split = date_only.split("/")

        day = date_split[1]
        month = date_split[0]
        year= date_split[2]

        rows_labels[idx]= year + "-%02d" % int(month) + "-%02d" % int(day)
    
df_usd2eur.info()
````

````python
# df2 is the df DataFrame with the USD to EUR values
df2 = pd.merge(df, df_usd2eur, on='Date')
df2.head(n=3)
````

## After the merge, the new column is Close_y, and Close is now Close_x.
    * Change Close_y to Close and Close_x to usd2eur

## To remove the days not present in the df DataFrame, df2 reveice itself filtered.

````python
close_x_index, = np.where(df2.columns.values == 'Close_x')
close_y_index, = np.where(df2.columns.values == 'Close_y')

# Condition to not change twice the columns names.
if close_x_index.size != 0:
    cols = df2.columns.values
    cols[close_y_index] = 'usd2eur'
    cols[close_x_index] = 'Close'
````

## Test data consistency in df2 DataFrame

````python
for i in range(df2.index.values.size):
    # If 'Close' from the row i in df2 is not the same in df.
    df_value = df.iloc[np.where(df.index.values == df2.index.values[i])]['Close'].values[0]
    if df2.iloc[i, 3] != df_value:
        print('Inconsistent Data in iloc: {}'.format(i))
        print('df2 value {}'.format(df2.iloc[i, 3]))
        print('df value {}'.format(df_value))
````
